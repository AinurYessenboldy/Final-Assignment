**This notebook contains data cleansing, data manupilation, and some data transformation operations on dataset about used cars of cities Chicago and LA. Firstly, I imported file with pandas, since it is a csv and tabular format works best for it. Then describe function is used to see some statistical summary about data.**

```python
import pandas as pd
file = pd.read_csv('used_cars_2cities.csv')
file[:100]
```

```python
describing_table = file.describe(include="all")
print(describing_table)
print("Now data types:\n",  file.dtypes)
```

**Operations below include removing first character from column price; then creating new column "year" by extracting it from column name. Next, several data transformation functions were performed on name, since it needed quite alterations.**

```python
file['price']=file['price'].str[1:]
file[:100]
```

```python
file['year']=file['name'].str[:4]
file[:100]
```

```python
file['name'] = file['name'].str.extract((r'(?i)(t.*)'), expand=False)
file['name'] = file['name'].str.lower()
file['name'] = file['name'].str.replace('_', '', regex=False)
file['name'] = file['name'].str.replace('-', '', regex=False)
file['name'] = file['name'].str.replace(r'\s+', ' ', regex=True)  # Remove extra spaces
file['name'] = file['name'].str.replace(r'\.', '', regex=True)    # Remove periods
file['name'] = file['name'].str.replace(r'\b\d{4}\b', '', regex=True)  # Remove 4-digit years
file['name'] = file['name'].str.strip()
file['name'] = file['name'].str.replace(r'bad credit', 'bad_credit', regex=True)
file['name'] = file['name'].str.replace(r'\s+4dr\s+', ' sedan ', regex=True)
file[:100]
```

**Below you can find dictionary from three columns, which is generated out of column names as key and their unique values are dictionary values. It is followed lambda formula applied on cylinders column to remove wording 'cylinders'. Next, change of data types takes place, altering some columns into integers and some to strings.**

```python
unique_values = {column: file[column].unique().tolist() for column in ['name', 'condition','cylinders']}
print(unique_values)
```

```python
file['cylinders'] = file['cylinders'].apply(lambda x: x.replace('cylinders', '').strip() if isinstance(x, str) else x)
file[:100]
```

```python
int_columns = ['price','year']
file[int_columns] = file[int_columns].fillna(0).astype(int)
string_columns = ['area', 'subarea','name','condition','drive','fuel','paintcolor','size','transmission','type']
file[string_columns] = file[string_columns].fillna(0).astype(str)
file[:200]
```

**This cell categorizes price into three groups according to its value. First, list of categories is given, then using for loop, each instance is checked to find to which category it belongs. Then number of obersations in each category is printed.**

```python
price_list = file['price'].tolist()
categories = ['cheap', 'average', 'expensive']
price_categories = []
for x in price_list:
    if x<5000:
        price_categories.append(categories[0])
    elif 5000 <= x <= 10000:
        price_categories.append(categories[1])
    else:
        price_categories.append(categories[2])
price_category_counts = pd.Series(price_categories).value_counts()
print(price_category_counts)
```

**Next, in order to understand distribution of conditions of cars, bar chart was plotted using matplotlib library.**

```python
import matplotlib.pyplot as plt
condition_counts = file['condition'].value_counts()
condition_counts.plot(kind='bar',color='pink')
plt.xlabel('Conditions')
plt.ylabel('Сount of сategories')
plt.title('Plot of different categories')
plt.show()
```

**Lastly, data is filtered to see instances not earlier than 2000, and only columns interesting to me are remained. Out of curiosity, new column with bitcoin price was generated, where I applied lambda formula of dividing it into 63, as of today's bitcoin worth to dollar.**

```python
filtered_data = file[file['year'] >= 2000]  
filtered_data = filtered_data[['name','year','price', 'condition']]
filtered_data['bitcoin_price'] = filtered_data['price'].apply(lambda x: (x/63) if x > 0 else x)
filtered_data
```
